```python
import multiprocessing as mp
from loguru import logger
from Bio import SeqIO 
from tqdm import tqdm
import biotite.sequence as seq
import biotite.sequence.align as align
from tqdm import tqdm

first_dict = SeqIO.to_dict(SeqIO.parse(open('sequence/MISATO.fasta'),'fasta')) # 直接转为字典格式
second_dict = SeqIO.to_dict(SeqIO.parse(open('sequence/lba+lep.fasta'),'fasta'))

pbar = tqdm(total=len(first_dict.keys()))
pbar.set_description('Sleep')
update = lambda *args: pbar.update()

matrix = align.SubstitutionMatrix.std_protein_matrix()

class work():
    def __init__(self,d1,d2,matrix):
        self.manager = mp.Manager
        self.ans = self.manager().list()
        self.d1 = d1
        self.d2 = d2
        self.d3 = matrix
 
    def finish_works(self,t):
        seq1 = seq.ProteinSequence(self.d1[t].seq)
        for t2 in self.d2:
            seq2 = seq.ProteinSequence(self.d2[t2].seq)
            alignments = align.align_optimal(seq1, seq2, matrix, local=False)
            if align.get_sequence_identity(alignments[0]) >= 0.3:
                self.ans.append(t[:4])
                return None

    def start(self):
        pool = mp.Pool(processes=24)
        for i in self.d1.keys():
            pool.apply_async(self.finish_works, (i,),callback=update)
        pool.close()
        pool.join()
wk = work(first_dict,second_dict,matrix)
wk.start()
```
